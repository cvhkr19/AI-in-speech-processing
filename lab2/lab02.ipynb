```python
from IPython.display import Audio
import librosa
```

```python
y, sr = librosa.load("ai in speech processing.wav")
```

```python
print(y)
print(sr)
```

```python
import numpy as np
```

```python
delF = np.zeros(len(y))
delF[0] = y[0]
for i in range(1,len(y)):
    delF[i] = y[i]-y[i-1]
print(delF)
```

```python
import librosa.display
librosa.display.waveshow(delF,color = 'blue')
```

```python
Audio(delF,rate = sr)
```

```python
zero_crossing_points = np.zeros(len(y))
for i in range(1,len(y)):
    if(y[i-1] == 0 and y[i] == 0):
        continue
    if(y[i-1]*y[i] <= 0):
        zero_crossing_points[i] = 1
num_zeros = []
first_one = False
count_zeros = 0
for i in range(len(y)):
    if(zero_crossing_points[i] == 0):
        count_zeros += 1
    else:
        num_zeros.append(count_zeros)
        count_zeros = 0
print(num_zeros)
```

```python
silence = num_zeros[0]/1
speech = sum(num_zeros[1:])/(len(num_zeros)-1)
print("average length (number of zeros in between) silence : ",silence)
print("avg length speech : ",speech)
```

```python
x = np.zeros(len(num_zeros))
for i in range(len(num_zeros)):
    x[i] = float(num_zeros[i])
librosa.display.waveshow(x)
```
